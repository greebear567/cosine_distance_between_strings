```python
import re
from collections import defaultdict
import numpy
from scipy.spatial.distance import cosine

L=[]
f=open("anaconda.txt", 'r')
for line in f:
    line=line.lower()
    some=re.split('[^a-z]', line)
    L.append([x for x in some if x])
f.close()
#создали список, элементы в нем списки, в которых элементы - слова строк

d=defaultdict(int)
j=0
for line in L:
    for word in line:
        if d[word]==0:
            d[word]+=j
            j+=1
#создали словарь - ключ=слово, значение=первое вхождение слова в текст

'''Для каждого слова определите его позицию в матрице Matrix, где

i - номер строки матрицы и номер строки исходного текста
j - номер столбца матрицы и номер ИНДЕКСА слова в словаре
И запишите в качестве элемента матрицы число вхождений слова в конкретное предложение
'''
arr=numpy.zeros((len(L),len(d)))
counter_line=0
for line in L:
    for word in line:
        arr[counter_line][d[word]]+=1
    counter_line+=1
k=0
for some in arr:
    for el in some:
        if el!=0:
            k+=el
print(k)#сумма всех элементов матрицы
print(len([i for line in L for i in line]))#количество элементов в списке списков

print('Введите номер первой строки')
first_str=int(input())
print('Введите номер второй строки')
second_str=int(input())
print('Первая строка: ', L[first_str])
print('Вторая строка: ', L[second_str])
print(arr[first_str])
print(arr[second_str])
od_sl=0
for i in range(len(arr[first_str])):
    if arr[first_str][i]!=0 and arr[second_str][i]!=0 :
        od_sl+=min(arr[first_str][i],arr[second_str][i])
        print('столбец совпадения: ', i)
print('число слов в первой строке', sum(i for i in arr[first_str]))
print('число слов во второй строке', sum(i for i in arr[second_str]))
print('число одинаковых слов =', od_sl)
somesome=cosine(arr[second_str],arr[first_str])
print('косинусное расстояние между ними', somesome)
```
